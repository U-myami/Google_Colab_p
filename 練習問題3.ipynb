```python
#お気に入りコスメリストの比較
#構造体Cosmeticを定義
#各データを作成してリストに格納
#条件に合うもの(categoriesがlip・ratingかつ4.5以上のもの)のみ抽出して出力


#構造体
from dataclasses import dataclass

#構造体Cosmeticsを定義
@dataclass
class Cosmetic:
  brand:str         #ブランド名
  name:str         #商品名
  category:str    #商品カテゴリ
  price:int        #価格
  rating:float     #評価(0.0~5.0)

#ブランド名　Against_Cherry_Blossom

#商品ラインナップ作成(5個)
cosmetics = [
    Cosmetic("Against_Cherry_Blossom", "Petal Kiss Tint", "lip", 3000, 4.6),
    Cosmetic("Against_Cherry_Blossom", "Sakura Glow Blush", "face", 4500, 4.5),
    Cosmetic("Against_Cherry_Blossom", "Blossom Veil Shadow", "eye", 5400, 4.1),
    Cosmetic("Against_Cherry_Blossom", "Cherry Dew Gloss", "lip", 2500, 5.0),
    Cosmetic("Against_Cherry_Blossom", "Hanami Silky Powder", "face", 3050, 2.5)
]

#categoriesがlip・ratingかつ4.5以上のものを抽出
top_lip = []
for item in cosmetics:
    if item.category == "lip" and item.rating >= 4.5:
        top_lip.append([ item.name, item.price])
print(top_lip)
```
