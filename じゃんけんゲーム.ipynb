```python
import random

# メイン処理
def janken():

  # じゃんけんの種類
  hands = {0:"グー",1:"チョキ",2:"パー"}

  # フラグ宣言
  keep_Playing = True   # あいこならループ
  is_hand_Cnt = True    # 入力エラーをチェック

  # ループ
  while keep_Playing:

    # フラグを初期化
    is_hand_Cnt = False

    # 0～2を入力
    print("あなたはどれを出す？ : 数値を入力")
    user_Hand_input = input("  0:グー ,1:チョキ ,2:パー => ")

    # 入力の数値チェック(数字かどうか、0から2の数字か)
    if not user_Hand_input.isdigit() or int(user_Hand_input) not in hands:
      print("正しい数値を入力してください")
      is_hand_Cnt = True # 入力エラーフラグを立てる
      continue # 以下の処理をスキップし、whileループの先頭に戻る

    # 入力した値を代入
    user_Hand = int(user_Hand_input)

    # CPU
    cpu_Hand = random.randint(0,2)

    print("あなた : ",hands[user_Hand])
    print("CPU    : ",hands[cpu_Hand],"\n")

    # じゃんけん勝負
    if user_Hand == cpu_Hand:
      print("引き分け\n")

    elif(user_Hand == 0 and cpu_Hand == 1) or \
        (user_Hand == 1 and cpu_Hand == 2) or \
        (user_Hand == 2 and cpu_Hand == 0):

      print("あなたの勝ち！！！\n")
      keep_Playing = False

    else:
      print("あなたの負け・・・\n")
      keep_Playing = False


# 関数呼び出し
janken()
```
