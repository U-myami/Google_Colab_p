```python
# 二人の学生管理システムのプログラムを作成してください。
# 1.学生の名前(10文字以内)、年齢(整数)、3科目のテストの点数(小数点OK)をメンバに持つ
# 　構造体Studentを定義してください。
# 2.学生のデータを入力し、一覧データを出力してください。
# 3.二人のテストの点数を比較して、科目ごとに点数が高い学生の名前と点数を配列に格納してください。
# 出力イメージ : [['谷野',80.7],['田中',70.9],['田中',98.4]]


#構造体
from dataclasses import dataclass

#構造体studentを定義(1)
@dataclass
class student:
  name : str      #名前→name
  age : int       #年齢→age
  point_Engrish :float
  point_Mathematics :float
  point_National_Language :float    #点数→point

#新規リストを作成(2)
name = ["冴月 麟","多々良 小傘"]
age = [17,18]
point_Engrish = [73.5,75.9]               #英語の点数
point_Mathematics = [75.2,70.4]           #数学の点数
point_National_Language = [80.9,89.7]     #国語の点数

#学生一覧
students = [
    student("冴月 麟", 17, 73.5, 75.2, 80.9),
    student("多々良 小傘", 18, 75.9, 70.4, 89.7)
]

#比較用リスト作成(3)の準備
RainToppres = []
#各科目(英・数・国)をまとめるための新規リスト作成
subjects = ["point_Engrish","point_Mathematics","point_National_Language"]

#2人分の点数を比較し科目毎に点数が高い学生の名前と点数を新規リスト内に格納(3)

#各ステータス作成時→ループ使用(forで1ステータスずつチェック)
#点数が高い学生の名前と科目をRainToppersのリストに格納(条件分岐使用)
for subject in subjects:
  if getattr(students[0],subject) > getattr(students[1],subject):            #点数が高い科目とその人の名前を比較
     RainToppres.append([students[0].name,getattr(students[0],subject)])     # RainToppresリストに格納
  else:
    RainToppres.append([students[1].name,getattr(students[1],subject)])      #低い場合は入れない

print(RainToppres)
```
