```python
from re import sub
# 二人の学生管理システムのプログラムを作成してください。
# 1.学生の名前(10文字以内)、年齢(整数)、3科目のテストの点数(小数点OK)をメンバに持つ
# 　構造体Studentを定義してください。
# 2.学生のデータを入力し、一覧データを出力してください。
# 3.二人のテストの点数を比較して、科目ごとに点数が高い学生の名前と点数を配列に格納してください。
# 出力イメージ : [['谷野',80.7],['田中',70.9],['田中',98.4]]


#構造体
from dataclasses import dataclass

#構造体studentを定義(1)
@dataclass
class student:
  name : str      #名前→name
  age : int       #年齢→age
  point :float    #点数→point

#新規リストを作成(2)
name = ["冴月 麟","多々良 小傘"]
age  = [17,18]
engrish_point = [73.5,75.9]                  #英語の点数
mathematics_point = [75.2,70.4]              #数学の点数
national_language_point = [80.9,89.7]         #国語の点数

#新規リスト作成(3)の準備
RainToppres = []

#2人分の点数を比較し科目毎に点数が高い学生の名前と点数を新規リスト内に格納(3)

#各科目(英・数・国)をまとめるための新規リスト作成
subjects = [engrish_point,mathematics_point,national_language_point]

#各ステータス作成時→ループ使用(forで1ステータスずつチェック)
#点数が高い学生の名前と科目をRainToppersのリストに格納(条件分岐使用)
for subjects in subjects:
  if subjects[0] > subjects[1]:
    RainToppres.append([name[0],subjects[0]])
  else:
    RainToppres.append([name[1],subjects[1]])

print(RainToppres)
```
