```python
from dataclasses import dataclass

# 構造体Studentを定義
@dataclass
class Student:
    name: str
    age: int
    point_English: float
    point_Mathematics: float
    point_National_Language: float


# 学生データ入力
students = []

for i in range(2):
    print(f"{i+1}人目の学生情報を入力してください")

    name = input("名前（10文字以内）: ")
    age = int(input("年齢: "))
    eng = float(input("英語の点数: "))
    math = float(input("数学の点数: "))
    jap = float(input("国語の点数: "))

    students.append(Student(name, age, eng, math, jap))
    print()

# 一覧表示
print("=== 学生一覧 ===")
for s in students:
    print(s)

# 科目ごとに点数が高い学生を比較
rain_Toppers = []
subjects = ["point_English", "point_Mathematics", "point_National_Language"]

for subject in subjects:
    if getattr(students[0], subject) > getattr(students[1], subject):
        rain_Toppers.append([students[0].name, getattr(students[0], subject)])
    else:
        rain_Toppers.append([students[1].name, getattr(students[1], subject)])

print("=== 各科目トップ ===")
print(rain_Toppers)
```
