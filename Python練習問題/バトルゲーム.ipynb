・バトルゲーム作成

[親作成](概念定義)
・クラスを使用

[子作成](親の概念を継承)
・キャラ設定(名前・HP・攻撃力・防御力)をinputで作成(int)
・各ターン開始時にHPとターン数を表示させる
・攻撃→相手が防御した場合はダメージ軽減(攻撃力-防御力=ダメージ)
・相手(CPU)の行動はランダム
・どちらかのHPが0以下で終了
・勝った方のキャラ名を表示


```python

#★バトル★

#[親作成](概念定義)
#・クラスを使用
class Character:
  def __init__(self, name, hp, attack, defense):
    self.name = name
    self.hp = hp
    self.attack = attack
    self.defense = defense

#[子作成](親の概念を継承)
#・キャラ設定(名前・HP・攻撃力・防御力)をinputで作成(int)
#・各ターン開始時にHPとターン数を表示させる
#・攻撃→相手が防御した場合はダメージ軽減(攻撃力-防御力=ダメージ)
#・相手(CPU)の行動はランダム
#・どちらかのHPが0以下で終了
#・勝った方のキャラ名を表示


#親クラスを継承してプレイヤーとCPUを作成
class Player(Character):
  pass
class CPU(Character):
  pass

import random                         #なぜここでimort randomが入るのかわからない

#プレイヤーのキャラ設定作成(inputで作成)
name = input("キャラ名: ")
hp = int(input("HP: "))
attack = int(input("攻撃力: "))
defense = int(input("防御力: "))

player = Player(name, hp, attack, defense)
cpu = CPU("NONAME", 300, 34, 28)     #調整かけていいかも

#バトル開始!どちらかのHPが0以下になるまで
turn = 1
while player.hp > 0 and cpu.hp > 0:

  #ターン数表示
  print("\n--- ターン", turn, "---")
  print(player.name, "HP:", player.hp)
  print(cpu.name, "HP:", cpu.hp)

  #プレイヤーの攻撃・防御をinputで決定
  #1=攻撃・2=防御
  #1・2以外の数字は反映させないようにする
  while True:
    action = input("行動を選択 (1:攻撃 2:防御): ")
    if action in ["1", "2"]:
        break
    print("1か2を入力してください")

  #CPUの挙動はランダム
  cpu_action = random.choice(["攻撃", "防御"])

  #プレイヤーのターン
  if action == "1":
    damage = player.attack - cpu.defense if cpu_action == "防御" else player.attack
    if damage < 0:
      damage = 0
    cpu.hp -= damage
    print(player.name, "の攻撃！", damage, "ダメージ")
  else:
    print(player.name, "は防御した")

  #CPUのターン
  if cpu_action == "攻撃":
    damage = cpu.attack - player.defense if action == "2" else cpu.attack
    if damage < 0:
      damage = 0
    player.hp -= damage
    print("CPUの攻撃！", damage, "ダメージ")
  else:
    print("CPUは防御した")

  turn += 1
#勝敗
if player.hp > 0:
  print("勝者:", player.name)
else:
  print("勝者: CPU")
```
