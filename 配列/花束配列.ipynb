```python
# 文字列に"紫"がある花は何もせずに格納、無い場合は"紫"を文字列の先頭に連結させて格納すること

# 花リスト
list_flowers = ["紫陽花","紫リンドウ","紫トルコキキョウ","レモンリーフ"]

# 格納用
list_new_flowers = []

# メイン処理
for flower in list_flowers:
  if "紫" in flower:                # "紫"がある
    pass
  else:                             # "紫"がない
    flower = "紫" + flower          # "紫" + 文字列

  list_new_flowers.append(flower)    # 格納


print(list_new_flowers)              # 出力
```
