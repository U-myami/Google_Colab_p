```python
# CSVにあるデータを追加してExcelに出力する
import csv

# 変数にフォルダパスを格納することで可読性や保守性を向上させる(リファクタリング)
# Widowsでファイルパスを指定するときはパスの前に"R"を記入すること!!
file_Path  = R"/content/drive/MyDrive/Colab Notebooks/素材/Sample.csv"

# 追記する(aは追記モード)
with open(file_Path, 'a') as f:
    writer = csv.writer(f)

    # indexも配列として格納
    writer.writerow(['7','島根','そば'])
    writer.writerow(['8','広島','もんじゃ焼き'])

```

```python
# CSVデータをExcelに出力する(unnamed: 0は非表示)
import csv
import os             # 今回はファイルパス結合に使用する
import pandas as pd

# 変数にフォルダパスを格納することで可読性や保守性を向上させる(リファクタリング)
# Widowsでファイルパスを指定するときはパスの前に"R"を記入すること!!
file_Path  = R"/content/drive/MyDrive/Colab Notebooks/素材/"

read_CsvPath = "Sample.csv"     # 読み込むCSV
excel_FilePath = "Sample.xlsx"  # 書き込むExcel

# ファイルパス結合(ダイレクトに結合しないこと)
read_CsvPath = os.path.join(file_Path, read_CsvPath)      # CSV
excel_FilePath = os.path.join(file_Path, excel_FilePath)  # Excel

print(read_CsvPath)
print(excel_FilePath)

# CSV読み込み(unnamed: 0を非表示にする)
data = pd.read_csv(read_CsvPath,index_col = 0)

# Excel形式で出力(Sheet nameで出力先のシート名を決める)
data.to_excel(excel_FilePath, sheet_name='食べ物',index = False)

```
