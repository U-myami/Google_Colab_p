```python
#お気に入りパフェ抽出
#構造体Parfaitを定義
#各データを作成してリストに格納
#評価が4.5以上のパフェの名前と価格を抽出して出力

from dataclasses import dataclass

#構造体Parfaitを定義
@dataclass
class Parfait:
 name : str       #パフェの名前
 fluit : str      #メインのフルーツ
 price : int      # 価格
 rating : float   # 評価

#5個のパフェを作ってリストに格納してみよう
parfaits = [
    Parfait("いちごパフェ","strawberry",1600,5.0),
    Parfait("無花果パフェ","fig",1600,4.5),
    Parfait("みかんパフェ","mandarin",950,2.5),
    Parfait("柿パフェ","persimmon",1500,3.7),
    Parfait("ぶどうパフェ","grapes",1300,3.2),
]

#格納用リスト作成
love_parfaits = []

#評価が4.5以上のパフェの名前と価格を抽出して出力
for love in parfaits:
   if love.rating >= 4.5:
    love_parfaits.append([love.name,love.price])
print(love_parfaits)
```
